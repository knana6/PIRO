```python
from transformers import pipeline

translator = pipeline(
    "translation",
    model="facebook/nllb-200-distilled-600M",
    src_lang="eng_Latn",
    tgt_lang="kor_Hang",
)

gen = pipeline("text-generation", model="gpt2")

out = gen("The future of AI is", max_new_tokens=100, do_sample=True, top_p=0.9)
text = out[0]["generated_text"]

translator(text)
```
